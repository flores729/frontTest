```python
import pandas as pd
import numpy as np

# 모든 \를 /로 바꿉니다.
with open("C:/Users/82106/Desktop/1회.csv", 'r', encoding='cp949') as file:
    lines = file.readlines()
    
df = pd.DataFrame(lines, columns=['입찰가격'])
df['입찰가격'] = df['입찰가격'].str.strip()
df['입찰가격'] = df['입찰가격'].str.replace(',', '')

# 현재 '입찰가격' 열은 문자열(object) 타입입니다.
print("--- 변환 전 ---")
print(df.info())

# .astype(int)를 사용하여 열 전체를 정수(int) 타입으로 변환합니다.
df['입찰가격'] = df['입찰가격'].astype(float)

# Dtype이 int64로 바뀐 것을 확인할 수 있습니다.
print("\n--- 변환 후 ---")
print(df.info())

print("\n--- 실제 데이터 ---")
print(df.head())
df.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.font_manager as fm

# 한글 폰트 설정 (이전과 동일)
try:
    font_path = fm.findfont(fm.FontProperties(family='Malgun Gothic'))
    if font_path:
        plt.rcParams['font.family'] = 'Malgun Gothic'
    else:
        font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'
        fm.fontManager.addfont(font_path)
        plt.rcParams['font.family'] = 'NanumGothic'
except:
    print("한글 폰트를 찾을 수 없어 기본 폰트로 설정합니다.")
plt.rcParams['axes.unicode_minus'] = False


# 1. 샘플 DataFrame 생성 (이미 'df'라는 DataFrame이 있다고 가정합니다)


# --- 여기부터가 핵심적인 코드입니다 ---

# 2. DataFrame에서 데이터 추출
# '입찰가격' 컬럼의 값들을 numpy 배열로 가져옵니다.
bids = df['입찰가격'].values 

# 우리 회사 입찰가 지정 (DataFrame의 값 중 하나를 선택)
our_company_bid = 1.342534e+10


# 3. 평균 계산 (Pandas 기능을 사용하면 더 편리합니다)
average_bid = df['입찰가격'].mean()

# 4. 시각화를 위한 인덱스 생성
company_indices = df.index.values # DataFrame의 인덱스를 그대로 사용
# 우리 회사의 인덱스를 찾습니다.
try:
    our_company_index = df[df['입찰가격'] == our_company_bid].index[0]
except IndexError:
    print("지정한 '우리 회사 입찰가'가 데이터에 존재하지 않습니다.")
    our_company_index = -1

# --- 이하 시각화 코드는 이전과 동일합니다 ---

# 5. 그래프 생성
fig, ax = plt.subplots(figsize=(12, 7))

# 전체 회사의 입찰가를 scatter plot으로 그립니다.
ax.scatter(company_indices, bids, label='타사 입찰가', c='royalblue', s=60)

# 우리 회사의 입찰가를 강조하여 그립니다.
if our_company_index != -1:
    ax.scatter(our_company_index, our_company_bid,
               label='자사 입찰가', c='red', s=150, ec='black', zorder=5)

# 6. 평균선 및 이격도 표시
ax.axhline(y=average_bid, color='green', linestyle='--', linewidth=2,
           label=f'평균 입찰가 ({average_bid:,.0f}원)')

if our_company_index != -1:
    deviation = our_company_bid - average_bid
    ax.vlines(x=our_company_index, ymin=average_bid, ymax=our_company_bid,
              color='red', linestyle=':', linewidth=2)
    ax.text(our_company_index + 0.1, average_bid + deviation / 2,
            f'차이:\n{deviation:,.0f}원', va='center', ha='left', fontsize=11, color='red')

# 7. 그래프 디자인 및 레이블 설정
ax.set_title('입찰 가격 분포 및 평균과의 이격도 분석 (DataFrame 활용)', fontsize=16, fontweight='bold')
ax.set_xlabel('입찰 회사 (DataFrame 인덱스)', fontsize=12)
ax.set_ylabel('입찰 가격 (원)', fontsize=12)
ax.grid(True, which='both', linestyle='--', linewidth=0.5)
ax.legend(fontsize=11)

ax.ticklabel_format(style='plain', axis='y')
ax.get_yaxis().set_major_formatter(
    plt.FuncFormatter(lambda y, _: f'{int(y):,}'))

ax.set_xticks(company_indices)
ax.set_xticklabels([i for i in company_indices])

plt.tight_layout()
plt.savefig('bid_analysis_df_plot.png')

print("DataFrame 기반 그래프가 'bid_analysis_df_plot.png' 파일로 저장되었습니다.")
```
